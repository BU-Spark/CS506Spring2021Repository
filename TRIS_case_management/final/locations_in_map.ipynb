```python
# read data
import pandas as pd


# courthouses
courthouse_data = pd.read_csv("./locations/Courthouses.csv")
courthouse_loc = courthouse_data['coordinates']
courthouse_latitude, courthouse_longitude = [], []
for i in range(len(courthouse_loc)):
    loc = courthouse_loc[i].split(", ")
    courthouse_latitude.append(float(loc[0]))
    courthouse_longitude.append(float(loc[1]))
print('Has ', len(courthouse_latitude), " courthouses.")

    
# YAD offices
YAD_data = pd.read_csv("./locations/YAD_offices.csv")
YAD_loc = YAD_data['coordinates']
YAD_latitude, YAD_longitude = [], []
for i in range(len(YAD_loc)):
    loc = YAD_loc[i].split(", ")
    YAD_latitude.append(float(loc[0]))
    YAD_longitude.append(float(loc[1]))
print('Has ', len(YAD_latitude), " YAD offices.")
    
    
# middle + high schools
school_data = pd.read_csv("./locations/Schools.csv")
school_loc = school_data['coordinates']
school_latitude, school_longitude = [], []
for i in range(len(school_loc)):
    loc = school_loc[i].split(", ")
    school_latitude.append(float(loc[0]))
    school_longitude.append(float(loc[1]))
print('Has ', len(school_latitude), " schools.")
```

```python
import folium

fmap = folium.Map(location=[42.361, -71.057], zoom_start=8)
# plot courthouses' locations
for i in range(len(courthouse_latitude)):
    folium.Circle(
        radius=2500,
        location=[courthouse_latitude[i], courthouse_longitude[i]],
        color="#555",
        weight=0,
        fill_color='#f54245', #red
        fill_opacity=0.8,
    ).add_to(fmap)
    
# plot YAD offices' locations
for i in range(len(YAD_latitude)):
    folium.Circle(
        radius=2000,
        location=[YAD_latitude[i], YAD_longitude[i]],
        color="#555",
        weight=0,
        fill_color='#4284f5', # blue
        fill_opacity=0.8,
    ).add_to(fmap)
    
# plot schools' locations
for i in range(len(school_latitude)):
    folium.Circle(
        radius=1000,
        location=[school_latitude[i], school_longitude[i]],
        color="#555",
        weight=0,
        fill_color='#d334eb', # purple
        fill_opacity=0.8,
    ).add_to(fmap)
    
fmap # show the map in notebook below
```
