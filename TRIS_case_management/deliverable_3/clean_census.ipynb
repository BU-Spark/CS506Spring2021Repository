```python
cols=[x for i,x in enumerate(tract.columns) if tract.iat[0,i]=='(X)']
tract.drop(columns=cols, inplace=True)
print(len(cols))
tract
```

```python
tract.to_csv('../data/census tract data/Education attainment/clean_1.csv', encoding='utf-8', index=None)
```

```python
no_sex = tract.copy()
for col in no_sex.columns[2:]:
    if 'Female' in col or 'Male' in col:
        del no_sex[col]

cols = {}
for col in no_sex.columns[2:]:
    splits = col.split('!!')
    new_name = '!!'.join(splits[3:]).replace('Population ', '')
    if 'percent' in col.lower():
        new_name = 'Percent!!' + new_name
    cols[col] = new_name

no_sex.rename(columns=cols, inplace=True)
no_sex
```

```python
no_sex.to_csv('../data/census tract data/Education attainment/clean_gender.csv', encoding='utf-8', index=None)
```

```python
only_percent = no_sex.copy()
for col in only_percent.columns[2:]:
    if not 'percent' in col.lower():
        del only_percent[col]

only_percent
```
