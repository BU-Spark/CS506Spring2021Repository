```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error
import seaborn as sns
```

```python
# dropping irrelevant data and non cooperative rounds
cols = list(range(2,21)) + list(range(42,62))
twins = pd.read_csv('Blake_RPD_Dataset-1_Twins-03-19-21.csv', header=0, usecols=cols, na_values=[' ',''])
twins = twins.dropna(axis=1, how='all')
twins = twins.dropna(axis=0, how='any')

# converting numerics
twins.iloc[:,:19] = twins.iloc[:,:19].astype(int)
i = 20
while i<39:
    twins.iloc[:,i] = twins.iloc[:,i].astype(float)
    i+=2

twins.head()

# subtract times
for index, row in twins.iterrows():
    new_times = [0, row['coop_rt1']]
    for i in range(2,11):
        new_times.append(row['coop_rt'+str(i)] - row['coop_rt'+str(i-1)])
    for i in range(2,11):
        twins.loc[index,'coop_rt'+str(i)] = new_times[i]
twins
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing

# splitting training and testing twins
twins_1 = pd.DataFrame(columns=twins.columns)
twins_2 = pd.DataFrame(columns=twins.columns)
for index, row in twins.iterrows():
    if row['Random_twin_split'] == 1:
        twins_1 = twins_1.append(row, ignore_index=True)
    else:
        twins_2 = twins_2.append(row, ignore_index=True)
        
twins_1.head()
twins_1.to_csv('dataset_with_strategies.csv',index = False)

le_train = preprocessing.LabelEncoder()
le_test = preprocessing.LabelEncoder()

X_train = twins_1.drop(columns=['Random_twin_split']+['coop'+str(i) for i in range(1,11)]\
                               +['coop_rt'+str(i) for i in range(1,11)])
X_test = twins_2.drop(columns=['Random_twin_split']+['coop'+str(i) for i in range(1,11)]\
                               +['coop_rt'+str(i) for i in range(1,11)])
```

```python
y1_train = le_train.fit_transform(twins_1['coop1'])
y1_test = le_test.fit_transform(twins_2['coop1'])

reg1 = LogisticRegression().fit(X_train, y1_train)

y2_train = le_train.fit_transform(twins_1['coop2'])
y2_test = le_test.fit_transform(twins_2['coop2'])

reg2 = LogisticRegression().fit(X_train, y2_train)

y3_train = le_train.fit_transform(twins_1['coop3'])
y3_test = le_test.fit_transform(twins_2['coop3'])

reg3 = LogisticRegression().fit(X_train, y3_train)

y4_train = le_train.fit_transform(twins_1['coop4'])
y4_test = le_test.fit_transform(twins_2['coop4'])

reg4 = LogisticRegression().fit(X_train, y4_train)

y5_train = le_train.fit_transform(twins_1['coop5'])
y5_test = le_test.fit_transform(twins_2['coop5'])

reg5 = LogisticRegression().fit(X_train, y5_train)

y6_train = le_train.fit_transform(twins_1['coop6'])
y6_test = le_test.fit_transform(twins_2['coop6'])

reg6 = LogisticRegression().fit(X_train, y6_train)

y7_train = le_train.fit_transform(twins_1['coop7'])
y7_test = le_test.fit_transform(twins_2['coop7'])

reg7 = LogisticRegression().fit(X_train, y7_train)

y8_train = le_train.fit_transform(twins_1['coop8'])
y8_test = le_test.fit_transform(twins_2['coop8'])

reg8 = LogisticRegression().fit(X_train, y8_train)

y9_train = le_train.fit_transform(twins_1['coop9'])
y9_test = le_test.fit_transform(twins_2['coop9'])

reg9 = LogisticRegression().fit(X_train, y9_train)

y10_train = le_train.fit_transform(twins_1['coop10'])
y10_test = le_test.fit_transform(twins_2['coop10'])

reg10 = LogisticRegression().fit(X_train, y10_train)
```

```python
print('trial 1 score:', reg1.score(X_test,y1_test))
print('trial 2 score:', reg2.score(X_test,y2_test))
print('trial 3 score:', reg3.score(X_test,y3_test))
print('trial 4 score:', reg4.score(X_test,y4_test))
print('trial 5 score:', reg5.score(X_test,y5_test))
print('trial 6 score:', reg6.score(X_test,y6_test))
print('trial 7 score:', reg7.score(X_test,y7_test))
print('trial 8 score:', reg8.score(X_test,y8_test))
print('trial 9 score:', reg9.score(X_test,y9_test))
print('trial 10 score:', reg10.score(X_test,y10_test))
```
