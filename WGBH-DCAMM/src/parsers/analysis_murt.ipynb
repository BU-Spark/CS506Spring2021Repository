```python
import pandas as pd
```

```python
df = pd.read_csv("../../data/WorkforceUtilizationSummaryReportApril2019_analysis_2.csv")

df.head(50)
```

```python
type_of_hire = ['Total  Journey Hours', 'Total Apprentice Hours','Total New Hire Hours']
```

```python
df.columns
```

```python
df.columns = ['craft_level', 'job_type', 'total_employee', 'caucasian',
       'african_american', 'hispanic', 'asian', 'native_american', 'other',
       'not_specified', 'total_female', 'total_male']
```

```python
df
```

```python
df = df[df['craft_level'].isin(type_of_hire)]
```

```python
df.columns = ['craft_level', 'total_employee', 'caucasian',
       'african_american', 'hispanic', 'asian', 'native_american', 'other',
       'not_specified', 'total_female', 'total_male', 'to_drop']
df = df.drop('to_drop', 1)
df
```

```python
df.info()
```

```python
df.columns[1:]
```

```python
for col in df.columns[1:]:
    df[col] = [float(str(i).replace(",", "")) for i in df[col]]
df
```

```python
df['craft_level'] = df['craft_level'].astype(str)
```

```python
df.info()
```

```python
df1 = df.groupby(['craft_level'], as_index=False).sum()
df1
```

```python
df1 = df1.T.reset_index()
df1
```

```python
df1.columns = df1.iloc[0].tolist()
df1 = df1.drop(0)
```

```python
df1
```

```python
# %matplotlib
# df1=df[:5]
%matplotlib inline
df1.plot('craft_level',kind = 'bar', figsize=(10,7))
```

```python
df2 = pd.concat([df1['craft_level'], df1[df1.columns.tolist()[1:]].sum(axis=1) ], axis=1)
df2.columns = ['craft_level', 'total_hours']
df2
```

```python
df2[:8]
```

```python
df2[:8].plot('craft_level',kind = 'line', figsize=(10,7))
```

```python
df3 = df2[8:].set_index('craft_level')
df3.columns = ['']
```

```python
df3.plot.pie(subplots=True, figsize=(11, 6))
```
