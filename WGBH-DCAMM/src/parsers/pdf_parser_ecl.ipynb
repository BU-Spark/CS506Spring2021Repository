```python
def hasLetter(inputString):
    return any(char.isalpha() for char in inputString)

def hasPoint(inputString):
    return any(char=='.' for char in inputString)
```

```python
# convert all PDFs in a directory
#import tabula
import tabula

def convert_str(s):
    types = [int, float]
    for t in types:
        try:
            return t(s)
        except ValueError:
            pass
    return str(s).strip('\"')
```

```python
# df2 = wrapper.read_pdf(
#         "./data/WorkforceUtilizationSummaryReportDec2019.pdf")

print()
### 
numbers = []
with open("WorkforceUtilizationSummaryReportDec2019.csv", 'r') as csv_file:
    lines = csv_file.readlines()
    i = 0
    for line in lines:
        row = []

        values = line.strip('\n')#.split(',')

        values = values.split(',')
        for val in values:
            if val is not '' and  not hasLetter(str(val)): #ADDED 
                if hasPoint(str(val)):
                    row.append(convert_str(val))

        numbers.append(row)


```
