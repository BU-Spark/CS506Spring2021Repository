```python
import pandas as pd 
import numpy as np 
import random 

data = pd.read_csv('filtered_final_dataset.csv')
data
```

```python
# Dictionary of the total amounts for each officer WITHOUT duplicate dates
total_sum_dic = {}
date_lst = {}
total_sum = 0
for n in data['Name'].unique():
    dates = []
    total_sum_dic[n] = 0
    for d in data.loc[data['Name'] == n, 'Date']:
        if d not in dates:
            dates.append(d)
            s = ((data.loc[(data['Date'] == d )& (data['Name'] == n), 'Amount'].values)[0])
            total_sum_dic[n] += s
    date_lst[n] = dates
    total_sum += total_sum_dic[n]
```

```python
date_lst_count = {}
for key, value in date_lst.items():
    date_lst_count[key] = len(value)
date_lst_count
```

```python
# Dictionary of the total misconducts for each officer WITHOUT duplicate case ID
total_dic = {}
misconduct_lst = {}
for n in data['Name'].unique():
    id = []
    total_dic[n] = []
    for d in data.loc[data['Name'] == n, 'CaseID']:
        if d not in id:
            id.append(d)
            s = ((data.loc[(data['CaseID'] == d )& (data['Name'] == n), 'TypeOfMisconduct'].values)[0])
            total_dic[n].append(s)
    misconduct_lst[n] = id
```

```python
num = {}
for key, value in misconduct_lst.items():
    num[key] = len(value)
num
```

```python
# Adding correct total contributions to dataset 
df = data.drop_duplicates(subset = ["Name"])
df['Total Amount']= df['Name'].map(total_sum_dic)
df['Total Contributions']= df['Name'].map(date_lst_count)
df['Total Misconducts'] = df['Name'].map(num)
df = df[['Name', 'Race','TypeOfMisconduct','Allegation', 'Date', 'Rank', 'Total Contributions', 'Total Amount', 'Total Misconducts']]
df
```

# Initial Exploration and Visualizations

```python
# Plot of Race of Officers Under Investigation
import matplotlib.pyplot as plt

df['Race'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Demographic of Officers Under Investigation")
plt.show()
```

```python
# Plot of Officer Rank 
df['Rank'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title(" Ranking of Officers Under Investigation")
plt.show()
```

```python
# Plot to see Demographic of Officers and the Contribution Amounts
import statistics

race1 = []
for race in df.Race:
    for amount in df.loc[df['Race'] == race, 'Total Amount']:
        if race == 'White':
            race1.append(amount)

race2 = []
for race in df.Race:
    for amount in df.loc[df['Race'] == race, 'Total Amount']:
        if race == 'Hispanic':
            race2.append(amount)

race3 = []
for race in df.Race[0:20]:
    for amount in df.loc[df['Race'] == race, 'Total Amount']:
        if race == 'Black':
            race3.append(amount)

med1 = statistics.median(race1)
med2 = statistics.median(race2)
med3 = statistics.median(race3)

print("Median Total Amount For Officers who are White: ", med1)
print("Median Total Amount For Officers that are Hispanic: ", med2)
print("Median Total Amount For Officers that are Black: ", med3)

race = ['white', 'Black', 'Hispanic']
amount = [med1, med3, med2]
plt.bar(race, amount)
plt.title("Median Contribution of Each Race of Officer Under Investigation")
plt.show()
```

```python
# Plot to see Demographic of Officers and the Amount of Contributions Made by Them
race1 = []
for race in df.Race:
    for amount in df.loc[df['Race'] == race, 'Total Contributions']:
        if race == 'White':
            race1.append(amount)

race2 = []
for race in df.Race:
    for amount in df.loc[df['Race'] == race, 'Total Contributions']:
        if race == 'Hispanic':
            race2.append(amount)

race3 = []
for race in df.Race[0:20]:
    for amount in df.loc[df['Race'] == race, 'Total Contributions']:
        if race == 'Black':
            race3.append(amount)

med1 = statistics.median(race1)
med2 = statistics.median(race2)
med3 = statistics.median(race3)

print("Median Amount of Contributions Made By Each Officers who are White: ", med1)
print("Median Amount of Contributions Made By Each Officers that are Hispanic: ", med2)
print("Median Amount of Contributions Made By Each Officers that are Black: ", med3)

race = ['white', 'Black', 'Hispanic']
amount = [med1, med3, med2]
plt.bar(race, amount)
plt.title("Median Amount of Contributions of Each Race of Officer Under Investigation")
plt.show()
```

```python
# Median amount of misconducts and the median amount of times they contributed 
complaint = []
for m in df.TypeOfMisconduct:
    for amount in df.loc[df['TypeOfMisconduct'] == m, 'Total Contributions']:
        if m == 'Citizen complaint':
            complaint.append(amount)
investigation = []
for i in df.TypeOfMisconduct:
    for amount in df.loc[df['TypeOfMisconduct'] == i, 'Total Contributions']:
        if i == 'Internal investigation':
            investigation.append(amount)
            
med1 = statistics.median(complaint)
med2 = statistics.median(investigation)

print("Median Amount of Contribution For Officers who's Misconduct is Citizen Complaint: ", med1)
print("Median Amount of Contribution For Officers who's Misconduct is Internal Investigation: ", med2)
```

```python
# Plot rank and the average contribution made by officer 

police = []
detective = []
sergeant = []
sergeant_detective = []
lieutenant = []
lieutenantenant_detective = []
captain = []
deputy_duperintendent = []

for i in df.Rank:
    for amount in df.loc[df['Rank'] == i, 'Total Contributions']:
        if i == 'Police Officer':
            police.append(amount)
        if i == 'Detective':
            detective.append(amount)
        if i == 'Sergeant':
            sergeant.append(amount)
        if i == 'Sergeant Detective':
            sergeant_detective.append(amount)
        if i == 'Lieutenant':
            lieutenant.append(amount)
        if i == 'Lieutenantenant Detective':
            lieutenantenant_detective.append(amount)
        if i == 'Captain':
            captain.append(amount)
        if i == 'Deputy Superintendent':
            deputy_duperintendent.append(amount)

med_police = statistics.median(police)
med_detective = statistics.median(detective)
med_sergeant = statistics.median(sergeant)
med_sergeant_detective = statistics.median(sergeant_detective)
med_lieutenant = statistics.median(lieutenant)
med_lieutenantenant_detective = statistics.median(lieutenantenant_detective)
med_captain = statistics.median(captain)
med_deputy_duperintendent = statistics.median(deputy_duperintendent)
            
print("Median Amount of Contribution For Police Officers: ", med_police)
print("Median Amount of Contribution For Detectives: ", med_detective)
print("Median Amount of Contribution For Sergeants: ", med_sergeant)
print("Median Amount of Contribution For Sergeant Detectives: ", med_sergeant_detective)
print("Median Amount of Contribution For Sergeant Lieutenants: ", med_lieutenant)
print("Median Amount of Contribution For Sergeant Lieutenantenant Detectives: ", med_lieutenantenant_detective)
print("Median Amount of Contribution For Captains: ", med_captain)
print("Median Amount of Contribution For Deputy Superintendents: ", med_deputy_duperintendent)

rank = ['Police Officer', 'Detective', 'Sergeant', 'Sergeant Detective', 'Lieutenant', 'Lieutenantenant Detective', 'Captain', 'Deputy Superintendent']
amount = [med_police, med_detective, med_sergeant, med_sergeant_detective, med_lieutenant, med_lieutenantenant_detective, med_captain, med_deputy_duperintendent]
plt.figure(figsize = (20,7))
plt.bar(rank, amount)
plt.title("Median Contribution of for Different Ranks")
plt.show()

# See how manhy officers are in which rank
# print(df['Rank'].value_counts())
```
