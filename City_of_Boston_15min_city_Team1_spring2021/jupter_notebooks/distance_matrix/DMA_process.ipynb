# Distance Matrix Request

```
import requests
```

```
def http_to_dic(httplinks):
  """
  input: 
    Distance Matrix API style http links
  
  output:
    dictionary
  """
  i=0
  dictionary = {}
  for link in httplinks:
    r = requests.get(url = link)
    data = r.json()
    dictionary[i] = data
    i+=1
  return dictionary
```

# <font color='red'><font size="+5">Change the httplinks[:10] to your index range. </font>

```
DMA_dict = http_to_dic(httplinks[:10])
```
