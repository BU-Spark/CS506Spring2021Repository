```python
import pandas as pd
import numpy as np
import folium

from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
geolocator = Nominatim(user_agent="get_parcel_coord")
```

```python
PATH = "/content/gdrive/My Drive/Colab Notebooks/CS506FinalProject/raw_parcels.csv"
```

```python
df = pd.read_csv(PATH)
df.head(10)
```

```python
# convert address to lat/lon
def get_coord(addr):

    # get the address with specific format with geolocator
    location = geolocator.geocode(addr, addressdetails=True, timeout=None)
    
    if location == None:
        return   
    else:
        raw = location.raw
        lat = float(raw['lat'])
        lon = float(raw['lon'])
    return lat, lon
```

```python
# put coordinates in the new column "coord" with the function get_coord
example = df.head(10)
example['coord'] = example['address_on_map'].apply(get_coord)
example.head(10)
```

```python
# split coordinates into latitude and longtitude
example[['lat', 'lon']] = pd.DataFrame(example['coord'].tolist(), index=example.index)
example = example.drop(columns=['address_on_map','coord'])
example.head(10)
```
