## Read in merged CSV files 

```python
import pandas as pd
import numpy as np

pres_data = pd.read_csv('MergedMassData.csv')

gov_data = pd.read_csv('MergedMassDataGov.csv')
```

```python
pres_data.columns
```

```python
pres_city = pres_data[['Tract', 'Total Population 2016', 'LatinX Population 2016',
       'Mexican 2016', 'Puerto Rican 2016', 'Cuban 2016', 'Other LatinX 2016',
       'Total Population 2019', 'LatinX Population 2019', 'Mexican 2019',
       'Puerto Rican 2019', 'Cuban 2019', 'Other LatinX 2019', 'City/Town',
       'Democratic 2016', 'Republican 2016', 'Total Votes Cast',
       'Democratic 2020', 'Republican 2020', 'Total Votes Cast 2020']]
```

```python
#convert Tract column to type str

pres_city['Tract'] = pres_city['Tract'].astype(str)
```

```python
pres_city.dtypes
```

```python
#Drop Tract Duplicates 
pres_city = pres_city.drop_duplicates(['Tract'])

#Sum up rows by Town 
pres_city = pres_city.groupby('City/Town').sum()
```

## Add columns with percent change

```python
pres_city['% LatinX 2016'] = pres_city['LatinX Population 2016']/pres_city['Total Population 2016']

pres_city['% LatinX 2019'] = pres_city['LatinX Population 2019']/pres_city['Total Population 2019']


pres_city = pres_city.rename(columns = {'Total Votes Cast':'Total Votes Cast 2016'})


pres_city['% Democratic 2016'] = pres_city['Democratic 2016']/pres_city['Total Votes Cast 2016']

pres_city['% Democratic 2020'] = pres_city['Democratic 2020']/pres_city['Total Votes Cast 2020']


pres_city['% Republican 2016'] = pres_city['Republican 2016']/pres_city['Total Votes Cast 2016']

pres_city['% Republican 2020'] = pres_city['Republican 2020']/pres_city['Total Votes Cast 2020']
```

## Add columns with percentage point change

```python
pres_city['Democratic Percentage point Change'] = (pres_city['% Democratic 2020'] - pres_city['% Democratic 2016'])*100

pres_city['Republican Percentage point Change'] = (pres_city['% Republican 2020'] - pres_city['% Republican 2016'])*100

pres_city['Change LatinX'] = (pres_city['% LatinX 2019'] - pres_city['% LatinX 2016'])*100
```

```python
#Latino pop shift overall

latinx_2016 = pres_city['LatinX Population 2016'].sum()

latinx_2019 = pres_city['LatinX Population 2019'].sum()


#total LatinX percentage increase in MA from 2016-2019
latinx_change = (latinx_2019 - latinx_2016) / latinx_2016

print(latinx_change*100)

total_pop_2016 = pres_city['Total Population 2016'].sum()

total_pop_2019 = pres_city['Total Population 2019'].sum()



latinx_percentage_2016 = latinx_2016/total_pop_2016

latinx_percentage_2019 = latinx_2019/total_pop_2019



#total LatinX percentage point increase in MA from 2016-2019
total_latinX_change = (latinx_percentage_2019 * 100) - (latinx_percentage_2016 * 100)
```

```python
#Percentage point change in Democratic support in MA as a whole from 2016-2020

Democratic_2016 = pres_city['Democratic 2016'].sum()

Democratic_2020 = pres_city['Democratic 2020'].sum()

total_votes_2016 = pres_city['Total Votes Cast 2016'].sum()

total_votes_2020 = pres_city['Total Votes Cast 2020'].sum()


Democratic_percentage_2016 = Democratic_2016/total_votes_2016

Democratic_percentage_2020 = Democratic_2020/total_votes_2020


#total Democratic percentage point change in MA from 2016-2019

total_democratic_change = (Democratic_percentage_2020*100) - (Democratic_percentage_2016*100)

print(total_democratic_change)
```

```python
#Percentage point change in Republican support in MA as a whole from 2016-2020


Republican_2016 = pres_city['Republican 2016'].sum()

Republican_2020 = pres_city['Republican 2020'].sum()

total_votes_2016 = pres_city['Total Votes Cast 2016'].sum()

total_votes_2020 = pres_city['Total Votes Cast 2020'].sum()


Republican_percentage_2016 = Republican_2016/total_votes_2016

Republican_percentage_2020 = Republican_2020/total_votes_2020


#total Democratic percentage point change in MA from 2016-2019

total_republican_change = (Republican_percentage_2020*100) - (Republican_percentage_2016*100)

print(total_republican_change)
```
