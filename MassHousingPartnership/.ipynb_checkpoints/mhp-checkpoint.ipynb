```python
import geopandas as gpd
```

```python
filename_ = "data/Statewide_parcels_SHP/L3_TAXPAR_POLY_ASSESS.shp"
```

```python
location_ = gpd.read_file(filename_)
```

```python
print(len(location_))
```

```python
sample_10_ = location_[:10]
sample_10_ = sample_10_.to_crs("EPSG:4326")
sample_10_
```

```python
sample_10_['area'] = sample_10_.area
sample_10_['centroid'] = sample_10_.centroid
```

```python
# print(type(sample_10_[0]))
print(sample_10_.info())
```

```python
sample_list_centroid_ = list(sample_10_['centroid'])
```

```python
print(type(sample_10_['NUM_ROOMS'][0]))
```

```python
def cal_centroid(sample_):
    X_ = []
    Y_ = []
    Weight_ = list(sample_['NUM_ROOMS'])
    for s in list(sample_['centroid']):
        X_.append(s.x)
        Y_.append(s.y)
    return X_, Y_, Weight_
```

```python
#1k
sample_1000_ = location_[:1000]
sample_1000_ = sample_1000_.to_crs("EPSG:4326")
sample_1000_['centroid'] = sample_1000_.centroid
X_, Y_, Weight_ = cal_centroid(sample_1000_)
```

```python
# 100k
sample_100000_ = location_[:100000]
sample_100000_ = sample_100000_.to_crs("EPSG:4326")
sample_100000_['centroid'] = sample_100000_.centroid
X_, Y_, Weight_ = cal_centroid(sample_100000_)
```

```python
# What about 1 million
sample_1000000_ = location_[:1000000]
sample_1000000_ = sample_1000000_.to_crs("EPSG:4326")
sample_1000000_['centroid'] = sample_1000000_.centroid
X_, Y_, Weight_ = cal_centroid(sample_1000000_)
```

```python
import matplotlib.pyplot as plt
import folium
from folium.plugins import HeatMap
from folium.plugins import MarkerCluster
```

```python
plt.plot(X_, Y_, 'o', color='black', markersize=1)
```

```python
def generateBaseMap(default_location = [42.3601, -71.0589]):
    base_map = folium.Map(location=default_location)
    return base_map
```

```python
base_map = generateBaseMap()
```

```python
HeatMap(
    data = [[lat_, long_, w_] for lat_, long_, w_ in zip(Y_, X_, Weight_)],
    radius = 12,
    max_zoom = 13).add_to(base_map)
```

```python
base_map.save('./maps/heat_map_1k.html')
```

```python
# Create scatter map
for x_, y_, w_ in zip(Y_, X_, Weight_):
    folium.Marker(
        location = [x_, y_]
    ).add_to(base_map)
```

```python
base_map.save('./maps/scatter_map_1k.html')
```
