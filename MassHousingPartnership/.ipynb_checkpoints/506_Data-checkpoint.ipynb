```python
small_quincy_parcels = quincy_parcels.iloc[0:100]
small_quincy_addresses = quincy_addresses.iloc[0:100]
m = []
for i, parcel in small_quincy_parcels.iterrows():
  counter = 0
  for j, address in small_quincy_addresses.iterrows():
    if address['geometry'].within(parcel['geometry']):
      counter += 1
  m.append(counter)
#small_quincy_parcels
small_quincy_parcels["addresses"]=m
```
