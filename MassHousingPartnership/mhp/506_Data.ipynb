```python
#Isolate quincy addresses
quincy_addresses = state_wide_address[state_wide_address['COMMUNITY_NAME']=='QUINCY']
quincy_addresses
```

```python
#Uploading Quincy parcel data (not a feather file)
quincy_parcels = gpd.read_file(from_geofeather(os.getcwd() + '/data/L3_SHP_M243_Quincy/M243TaxPar.shp')
quincy_parcels
#quincy_parcels["geometry"][0]
```

```python
#Installing Shapley. Shapley can help us find how many addresses are in each parcel via Point (addresses) in Polygon (parcel)
from shapely.geometry import Point, Polygon
```

```python
#Test lines to understand the datasets and shapley
parcel = quincy_parcels["geometry"][0]
type(parcel)
address = quincy_addresses["geometry"].iloc[0]
type(address)
address.within(parcel)
#POINT.winthin(POLYGON) returns true if point is in the polygon, and false if its not in the polygon

```

```python
#Proof of concept! It works :D
small_quincy_parcels = quincy_parcels.iloc[0:10]
small_quincy_addresses = quincy_addresses.iloc[0:100]

density = []
for i, parcel in small_quincy_parcels.iterrows():
  counter = 0
  for j, address in quincy_addresses.iterrows():
    if address['geometry'].within(parcel['geometry']):
      counter += 1
  density.append(counter)
print(density)
```

```python
newQuincy = gpd.sjoin(quincy_addresses, quincy_parcels[['LOC_ID', 'geometry']], how='left', op='within')
newQuincy
```

```python
#Calculating density for all of Quincy
density = []
for i, parcel in quincy_parcels.iterrows():
  counter = 0
  for j, address in quincy_addresses.iterrows():
    if address['geometry'].within(parcel['geometry']):
      counter += 1
  density.append(counter)
print(density)
```
